# Ensemble learning:-
 Ensemble learning is a technique in which we train so many model on our given dataset.And the small model which train on the training dataset called week model.Now whatever sample we get will predict there class using the combination of week models.

 # Strong model:- 
 these small model collectively represent a strong model.


```python
# Ensemble(group) learning techniques:

# Bagging

# Note: Bagging all algos create week model and data created using bootstrap
# Note: In Bootstrap same number of sample come in the training of week model and each weak model may contain duplicat sample manytime.
# 1> Randomforest regression

# 2> Randomforest classifier

# 3> BaggingClassifier

# 4> BaggingRegressor

# etc

# Boosting

# 1> AdaBoost Classifier

# 2> AdaBoost Regressor

# etc

# Stacking

# 1> VotingClassifier

# 2> VotingRegressor

# etc
```
